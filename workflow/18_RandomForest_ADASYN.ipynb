# Random Forest with ADASYN

The final notebook in this project's model comparison series. Given ADASYN is a
refinement of SMOTE, the expectation is that it lands close to SMOTE's
precision-favoring result rather than UnderSampling's recall-favoring one.

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, accuracy_score, precision_score,
                              f1_score, recall_score, confusion_matrix,
                              ConfusionMatrixDisplay, roc_auc_score,
                              average_precision_score, roc_curve, precision_recall_curve)
from imblearn.over_sampling import ADASYN
from sklearn.compose import ColumnTransformer
from imblearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load Data

```python
df = pd.read_csv('03_BankCustomer_OutlierChecked.csv')
df.head()
```

## 3. Define Features and Target

```python
X = df.drop(columns=['CustomerId', 'Surname', 'Exited'])
y = df['Exited']
```

## 4. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## 5. Preprocessing

```python
num_features = ["CreditScore", "Age", "EstimatedSalary", "Balance", "NumOfProducts", "Tenure"]
cat_features = ["Geography", "Gender", "HasCrCard", "IsActiveMember"]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_features)
    ]
)
```

## 6. Build the Pipeline with ADASYN

```python
steps = [("preprocess", preprocessor),
         ("adasyn", ADASYN(random_state=42)),
         ("random_forest", RandomForestClassifier(random_state=42))]

pipe = Pipeline(steps)
pipe
```

## 7. Baseline: Untuned Forest + ADASYN

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
y_pred
```

## 8. Hyperparameter Grid

```python
param_grid = {
    'random_forest__n_estimators': [100, 200, 300],
    'random_forest__max_depth': [10, 20, None],
    'random_forest__min_samples_split': [2, 5, 10],
    'random_forest__min_samples_leaf': [1, 2, 4],
    'random_forest__class_weight': [None, 'balanced']
}
```

## 9. Run Grid Search

```python
grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5,
    scoring='recall',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

## 10. Predict Using the Best Model

```python
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)
y_pred
```

## 11. Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Churn", "Churn"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix — Random Forest + ADASYN")
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 12. Per-Class Recall

```python
class0_recall = cm[0, 0] / cm[0].sum()
class1_recall = cm[1, 1] / cm[1].sum()

print(f"Class 0 (Not Churn) Recall: {class0_recall:.2f}")
print(f"Class 1 (Churn) Recall: {class1_recall:.2f}")
```

## 13. Overall Metrics

```python
print(f'Accuracy  : {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision : {precision_score(y_test, y_pred):.2f}')
print(f'Recall    : {recall_score(y_test, y_pred):.2f}')
print(f'F1 Score  : {f1_score(y_test, y_pred):.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

## 14. Predicted Probabilities and Threshold-Independent Metrics

```python
y_prob = best_rf.predict_proba(X_test)[:, 1]
y_prob
```

```python
roc_auc = roc_auc_score(y_test, y_prob)
pr_auc = average_precision_score(y_test, y_prob)
print(f"ROC-AUC Score: {roc_auc:.2f}")
print(f"PR-AUC Score : {pr_auc:.2f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, linewidth=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--", linewidth=1)
plt.xlabel("False Positive Rate", fontsize=12)
plt.ylabel("True Positive Rate", fontsize=12)
plt.title("ROC Curve — Random Forest + ADASYN", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(recall, precision, linewidth=2, marker=".", label=f"PR Curve (AUC = {pr_auc:.2f})")
plt.xlabel("Recall", fontsize=12)
plt.ylabel("Precision", fontsize=12)
plt.title("Precision-Recall Curve — Random Forest + ADASYN", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## 17. Project Conclusion

- **ADASYN confirms the prediction:** it lands close to SMOTE (Precision 0.62 vs 0.64,
  Recall 0.61 vs 0.60) — both meaningfully more conservative than UnderSampling, since
  both generate synthetic points via the same interpolation logic.
- **Best overall model: Random Forest with `class_weight='balanced'`** — highest
  ROC-AUC (0.88), highest PR-AUC (0.71), and the best F1 (0.65) of every model and
  technique tested across this entire project.
- **Three clear "profiles" among Random Forest variants**, useful for different business
  needs:
  - **`class_weight` alone** — best all-around balance
  - **UnderSampling** — best when missing a churner is costlier than a false alarm
  - **SMOTE/ADASYN** — best when false alarms are costlier than missing some churners
- **Cross-model-family lesson:** the "best" imbalance technique is not universal — it
  depends on the model. Linear models (Logistic Regression) were largely indifferent to
  which technique was used, since their real limitation was the model itself. Tree-based
  models responded very differently to each technique, revealing genuine, usable
  trade-offs rather than one clear winner.

**Final recommendation:** deploy the tuned Random Forest with `class_weight='balanced'`
as the primary churn model, and consider the UnderSampling variant specifically for a
higher-stakes retention campaign where catching more at-risk customers outweighs the
cost of a few extra false alarms.
