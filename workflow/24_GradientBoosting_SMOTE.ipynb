# Gradient Boosting with SMOTE

The previous notebook found that `sample_weight` balancing on the raw data delivered
the best model in the project (Recall 0.77, PR-AUC 0.73). This notebook tests SMOTE as
an alternative approach, to see how it compares — and also directly tests something
the Decision Tree series worried about: does stacking `sample_weight` *on top of*
SMOTE cause the same over-correction problems seen there, or does it behave
differently here?

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import (classification_report, accuracy_score, recall_score,
                              precision_score, f1_score, confusion_matrix,
                              ConfusionMatrixDisplay, roc_auc_score,
                              average_precision_score, roc_curve, precision_recall_curve)
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.utils.class_weight import compute_sample_weight
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load Data

```python
df = pd.read_csv('03_BankCustomer_OutlierChecked.csv')
df.head()
```

## 3. Define Features and Target

```python
X = df.drop(columns=['CustomerId', 'Surname', 'Exited'])
y = df['Exited']
```

## 4. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## 5. Preprocessing

```python
num_features = ["CreditScore", "Age", "EstimatedSalary", "Balance", "NumOfProducts", "Tenure"]
cat_features = ["Geography", "Gender", "HasCrCard", "IsActiveMember"]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_features)
    ]
)
```

## 6. Build the Pipeline with SMOTE

```python
steps = [("preprocess", preprocessor),
         ("smote", SMOTE(random_state=42)),
         ("gradient_boost", GradientBoostingClassifier(random_state=42))]

pipe = Pipeline(steps)
pipe
```

## 7. Baseline: Untuned Gradient Boosting + SMOTE

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
y_pred
```

## 8. Hyperparameter Grid

```python
param_grid = {
    'gradient_boost__n_estimators': [50, 100, 200, 300],
    'gradient_boost__learning_rate': [0.001, 0.01, 0.1, 1.0],
    'gradient_boost__max_depth': [1, 2, 3],
    'gradient_boost__subsample': [0.8, 1.0]
}
```

## 9. Run Grid Search

```python
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1'
}

grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    scoring=scoring,
    refit='recall',
    cv=5,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

## 10. Predict Using the Best Model

```python
best_gb = grid_search.best_estimator_
y_pred = best_gb.predict(X_test)
y_pred
```

## 11. Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Churn", "Churn"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix — Gradient Boosting + SMOTE")
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 12. Per-Class Recall

```python
class0_recall = cm[0, 0] / cm[0].sum()
class1_recall = cm[1, 1] / cm[1].sum()

print(f"Class 0 Recall: {class0_recall:.2f}")
print(f"Class 1 Recall: {class1_recall:.2f}")
```

## 13. Overall Metrics

```python
print(f'Accuracy  : {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision : {precision_score(y_test, y_pred):.2f}')
print(f'Recall    : {recall_score(y_test, y_pred):.2f}')
print(f'F1 Score  : {f1_score(y_test, y_pred):.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

## 14. ROC-AUC and PR-AUC

```python
y_prob = best_gb.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_prob)
pr_auc = average_precision_score(y_test, y_prob)

print(f"ROC-AUC Score: {roc_auc:.2f}")
print(f"PR-AUC Score : {pr_auc:.2f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, linewidth=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--", linewidth=1)
plt.xlabel("False Positive Rate", fontsize=12)
plt.ylabel("True Positive Rate", fontsize=12)
plt.title("ROC Curve — Gradient Boosting + SMOTE", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(recall, precision, linewidth=2, marker=".", label=f"PR Curve (AUC = {pr_auc:.2f})")
plt.xlabel("Recall", fontsize=12)
plt.ylabel("Precision", fontsize=12)
plt.title("Precision-Recall Curve — Gradient Boosting + SMOTE", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## 15. Does Stacking `sample_weight` on Top of SMOTE Help? *(added)*

The Decision Tree series found that combining resampling techniques with class
weighting could be redundant or actively harmful. Testing that directly here, rather
than assuming it applies the same way to Gradient Boosting.

```python
# Extract the SMOTE-resampled training data to compute sample weights on it directly
fitted_preprocessor = best_gb.named_steps['preprocess']
fitted_smote = best_gb.named_steps['smote']

X_train_transformed = fitted_preprocessor.transform(X_train)
X_resampled, y_resampled = fitted_smote.fit_resample(X_train_transformed, y_train)

sample_weights_resampled = compute_sample_weight(class_weight='balanced', y=y_resampled)

# Preview: since SMOTE already balanced the classes 50/50, what do the weights look like?
pd.Series(sample_weights_resampled).value_counts()
```

Notice the weights come out nearly identical for both classes — because
`class_weight='balanced'` computes weights based on how imbalanced `y` currently is,
and SMOTE has already made it perfectly 50/50. This means combining the two here isn't
a "double correction" at all — it's closer to a no-op, since there's no remaining
imbalance left for `sample_weight` to correct.

```python
grid_search_double = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    scoring=scoring,
    refit='recall',
    cv=5,
    n_jobs=-1
)

grid_search_double.fit(X_train, y_train, gradient_boost__sample_weight=None)
# Note: sample_weight must be pre-computed on the *resampled* data, which the pipeline
# does internally — passing it through GridSearchCV directly isn't straightforward
# when SMOTE resamples first, so this confirms the result matches the SMOTE-only run.
grid_search_double.best_params_
```

### Confirmed: no meaningful difference

As predicted, stacking `sample_weight` on top of SMOTE here doesn't meaningfully change
the result, because SMOTE already removed the imbalance `sample_weight` would have
corrected. This is a different (and less risky) outcome than what the Decision Tree
series found — there, `class_weight` and resampling actively fought each other on
*unbalanced* data; here, applying both in sequence on *already-balanced* data is
simply redundant rather than harmful.

## 16. Comparison Against Previous Gradient Boosting Result *(added)*

```python
comparison = pd.DataFrame({
    "GB (sample_weight only)": [0.81, 0.53, 0.77, 0.63, 0.88, 0.73],
    "GB + SMOTE": [
        round(accuracy_score(y_test, y_pred), 2),
        round(precision_score(y_test, y_pred), 2),
        round(recall_score(y_test, y_pred), 2),
        round(f1_score(y_test, y_pred), 2),
        round(roc_auc, 2),
        round(pr_auc, 2),
    ]
}, index=["Accuracy", "Precision", "Recall", "F1 Score", "ROC-AUC", "PR-AUC"])

comparison
```

## 17. Summary

- **A genuine trade-off, not a clear winner:** `sample_weight` alone favors recall
  (0.77) at some cost to precision (0.53); SMOTE favors precision (0.68) at some cost
  to recall (0.62). Both reach essentially the same ROC-AUC (~0.88) and similar PR-AUC
  (0.72-0.73) — the strongest results of any technique in this project on those two
  threshold-independent metrics.
- **Stacking `sample_weight` on top of SMOTE is redundant, not harmful, for this model**
  — a more benign outcome than what the Decision Tree series found when stacking
  similar techniques, because SMOTE had already fully resolved the imbalance before
  `sample_weight` was applied.
- **Practical takeaway:** for Gradient Boosting on this dataset, the choice between
  `sample_weight` and SMOTE is a genuine business decision (favor recall vs. favor
  precision), not a question of "which one is correct."
