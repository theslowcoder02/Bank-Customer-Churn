# AdaBoost with ADASYN

The final notebook in the AdaBoost series. Given ADASYN is a refinement of SMOTE, the
expectation — based on the previous notebook's finding — is that the shallow stump will
again outperform the depth-5 base learner, rather than repeating the pattern seen with
`class_weight`.

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (classification_report, accuracy_score, recall_score,
                              precision_score, f1_score, confusion_matrix,
                              ConfusionMatrixDisplay, roc_auc_score,
                              average_precision_score, roc_curve, precision_recall_curve)
from imblearn.over_sampling import ADASYN
from imblearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load Data

```python
df = pd.read_csv('03_BankCustomer_OutlierChecked.csv')
df.head()
```

## 3. Define Features and Target

```python
X = df.drop(columns=['CustomerId', 'Surname', 'Exited'])
y = df['Exited']
```

## 4. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## 5. Preprocessing

```python
num_features = ["CreditScore", "Age", "EstimatedSalary", "Balance", "NumOfProducts", "Tenure"]
cat_features = ["Geography", "Gender", "HasCrCard", "IsActiveMember"]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_features)
    ]
)
```

## 6. Build the Pipeline with ADASYN (Default Stump)

```python
steps = [("preprocess", preprocessor),
         ("adasyn", ADASYN(random_state=42)),
         ("ada_boost", AdaBoostClassifier(random_state=42))]

pipe = Pipeline(steps)
pipe
```

## 7. Baseline: Untuned AdaBoost + ADASYN

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
y_pred
```

## 8. Hyperparameter Grid

```python
param_grid = {
    'ada_boost__n_estimators': [50, 100, 200, 300],
    'ada_boost__learning_rate': [0.001, 0.01, 0.1, 1.0]
}
```

## 9. Run Grid Search

```python
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1'
}

grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    scoring=scoring,
    refit='recall',
    cv=5,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

## 10. Predict Using the Best Model

```python
best_ada = grid_search.best_estimator_
y_pred = best_ada.predict(X_test)
y_pred
```

## 11. Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Churn", "Churn"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix — AdaBoost + ADASYN (stump)")
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 12. Per-Class Recall

```python
class0_recall = cm[0, 0] / cm[0].sum()
class1_recall = cm[1, 1] / cm[1].sum()

print(f"Class 0 Recall: {class0_recall:.2f}")
print(f"Class 1 Recall: {class1_recall:.2f}")
```

## 13. Overall Metrics

```python
print(f'Accuracy  : {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision : {precision_score(y_test, y_pred):.2f}')
print(f'Recall    : {recall_score(y_test, y_pred):.2f}')
print(f'F1 Score  : {f1_score(y_test, y_pred):.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

## 14. ROC-AUC and PR-AUC

```python
y_prob = best_ada.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_prob)
pr_auc = average_precision_score(y_test, y_prob)

print(f"ROC-AUC Score: {roc_auc:.2f}")
print(f"PR-AUC Score : {pr_auc:.2f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, linewidth=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--", linewidth=1)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve — AdaBoost + ADASYN (stump)", fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(recall, precision, linewidth=2, marker=".", label=f"PR Curve (AUC = {pr_auc:.2f})")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve — AdaBoost + ADASYN (stump)", fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

## 15. Confirming the Pattern: Depth-5 with ADASYN *(added)*

Testing the same prediction made at the top of this notebook — that ADASYN, like
SMOTE, will *not* benefit from the deeper base learner.

```python
steps_depth5 = [("preprocess", preprocessor),
                ("adasyn", ADASYN(random_state=42)),
                ("ada_boost", AdaBoostClassifier(
                    random_state=42,
                    estimator=DecisionTreeClassifier(max_depth=5, random_state=42)
                ))]

pipe_depth5 = Pipeline(steps_depth5)

grid_search_depth5 = GridSearchCV(
    estimator=pipe_depth5,
    param_grid=param_grid,
    scoring=scoring,
    refit='recall',
    cv=5,
    n_jobs=-1
)

grid_search_depth5.fit(X_train, y_train)
grid_search_depth5.best_params_
```

```python
best_ada_depth5 = grid_search_depth5.best_estimator_
y_pred_depth5 = best_ada_depth5.predict(X_test)
y_prob_depth5 = best_ada_depth5.predict_proba(X_test)[:, 1]

print(f'Accuracy  : {accuracy_score(y_test, y_pred_depth5):.2f}')
print(f'Precision : {precision_score(y_test, y_pred_depth5):.2f}')
print(f'Recall    : {recall_score(y_test, y_pred_depth5):.2f}')
print(f'F1 Score  : {f1_score(y_test, y_pred_depth5):.2f}')
print(f'ROC-AUC   : {roc_auc_score(y_test, y_prob_depth5):.2f}')
print(f'PR-AUC    : {average_precision_score(y_test, y_prob_depth5):.2f}')
```

### Prediction confirmed

The shallow stump again outperforms the depth-5 base learner (~0.74 recall vs ~0.61),
mirroring the SMOTE notebook exactly. This gives a clear, consistent rule for AdaBoost
on this dataset — not a one-off coincidence.

## 17. AdaBoost Series Conclusion

- **Confirmed rule for this dataset:** `class_weight` pairs best with a deeper
  (depth=5) base learner; SMOTE and ADASYN both pair best with the default shallow
  stump. This held consistently across two separate resampling techniques, not just
  once.
- **Best AdaBoost configuration overall:** depth-5 + `class_weight`, Recall ~0.79,
  ROC-AUC ~0.87 — competitive with, but not quite matching, Random Forest's ROC-AUC
  (0.88) and PR-AUC (0.71).
- **Consistent project-wide theme:** `class_weight` continues to be the strongest or
  most reliable imbalance technique across every model family tested so far (Logistic
  Regression, Decision Tree, Random Forest, AdaBoost) — though *how* it's best applied
  (fixed vs. tuned, shallow vs. deep base learner) varies meaningfully by model.

**Running project ranking:** Random Forest (plain, class_weight) remains the top
overall model. AdaBoost (depth=5, class_weight) is the strongest boosting-based model
and the best challenger on Recall specifically. The next notebook — Gradient Boosting —
tests a more refined sequential boosting approach to see if it can finally close the
gap with, or surpass, Random Forest.
