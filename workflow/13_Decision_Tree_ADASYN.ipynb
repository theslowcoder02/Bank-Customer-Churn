# Decision Tree with ADASYN

This is the fourth and final imbalance-handling technique tested with Decision Trees.
Given that SMOTE (its close cousin) underperformed the plain tuned tree, the expectation
here is that ADASYN — which generates even more synthetic points around the "hardest"
churn cases — will show the same weakness, possibly more pronounced.

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (classification_report, accuracy_score, precision_score,
                              f1_score, recall_score, confusion_matrix,
                              ConfusionMatrixDisplay, roc_auc_score,
                              average_precision_score, roc_curve, precision_recall_curve)
from imblearn.over_sampling import ADASYN
from sklearn.compose import ColumnTransformer
from imblearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load Data

```python
df = pd.read_csv('03_BankCustomer_OutlierChecked.csv')
df.head()
```

## 3. Define Features and Target

```python
X = df.drop(columns=['CustomerId', 'Surname', 'Exited'])
y = df['Exited']
```

## 4. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## 5. Preprocessing

```python
num_features = ["CreditScore", "Age", "EstimatedSalary", "Balance", "NumOfProducts", "Tenure"]
cat_features = ["Geography", "Gender", "HasCrCard", "IsActiveMember"]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_features)
    ]
)
```

## 6. Build the Pipeline with ADASYN

```python
steps = [("preprocess", preprocessor),
         ("adasyn", ADASYN(random_state=42)),
         ("decision_tree", DecisionTreeClassifier(random_state=42))]

pipe = Pipeline(steps)
pipe
```

## 7. Baseline: Untuned Tree + ADASYN

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
y_pred
```

## 8. Hyperparameter Grid

```python
param_grid = {
    'decision_tree__criterion': ['gini', 'entropy'],
    'decision_tree__max_depth': [None, 10, 20, 30],
    'decision_tree__min_samples_split': [2, 5, 10],
    'decision_tree__min_samples_leaf': [1, 2, 4],
    'decision_tree__class_weight': [None, 'balanced']
}
```

## 9. Run Grid Search

```python
grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5,
    scoring='recall',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

## 10. Predict Using the Best Model

```python
best_dt = grid_search.best_estimator_
y_pred = best_dt.predict(X_test)
y_pred
```

## 11. Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Churn", "Churn"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix — Decision Tree + ADASYN")
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 12. Per-Class Recall

```python
class0_recall = cm[0, 0] / cm[0].sum()
class1_recall = cm[1, 1] / cm[1].sum()

print(f"Class 0 (Not Churn) Recall: {class0_recall:.2f}")
print(f"Class 1 (Churn) Recall: {class1_recall:.2f}")
```

## 13. Overall Metrics

```python
print(f'Accuracy  : {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision : {precision_score(y_test, y_pred):.2f}')
print(f'Recall    : {recall_score(y_test, y_pred):.2f}')
print(f'F1 Score  : {f1_score(y_test, y_pred):.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

## 14. Predicted Probabilities and Threshold-Independent Metrics

```python
y_prob = best_dt.predict_proba(X_test)[:, 1]
y_prob
```

```python
roc_auc = roc_auc_score(y_test, y_prob)
pr_auc = average_precision_score(y_test, y_prob)
print(f"ROC-AUC Score: {roc_auc:.2f}")
print(f"PR-AUC Score : {pr_auc:.2f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, linewidth=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--", linewidth=1)
plt.xlabel("False Positive Rate", fontsize=12)
plt.ylabel("True Positive Rate", fontsize=12)
plt.title("ROC Curve — Decision Tree + ADASYN", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(recall, precision, linewidth=2, marker=".", label=f"PR Curve (AUC = {pr_auc:.2f})")
plt.xlabel("Recall", fontsize=12)
plt.ylabel("Precision", fontsize=12)
plt.title("Precision-Recall Curve — Decision Tree + ADASYN", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## 16.  Conclusion — Decision Tree Series

**Overall takeaway across both model families:** the "best" way to handle class
imbalance is not universal — it depends on the model. `class_weight='balanced'`
worked well for both Logistic Regression and Decision Trees, and required no external
library, no synthetic data, and no discarded rows. External resampling techniques
(UnderSampling, SMOTE, ADASYN) added complexity without adding benefit for either
model type tested in this project — and actively hurt performance for Decision Trees.

**Best model overall so far:** Decision Tree with `class_weight='balanced'`, Recall
0.74, PR-AUC 0.59 — ahead of every Logistic Regression variant. The natural next step
is an ensemble method (e.g. Random Forest), to see whether combining many trees can
push performance higher still, or whether a single well-tuned tree has already
captured most of the available signal in this dataset.
