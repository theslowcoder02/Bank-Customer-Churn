# Random Forest with SMOTE

So far: the plain class-weighted Random Forest balances precision and recall well
(0.61 / 0.68); UnderSampling pushed recall much higher at some cost to precision
(0.49 / 0.78). This notebook tests SMOTE — generating synthetic churn examples instead
of removing real data — to see where it lands on that same spectrum.

```python
## 1. Import Libraries
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, accuracy_score, precision_score,
                              f1_score, recall_score, confusion_matrix,
                              ConfusionMatrixDisplay, roc_auc_score,
                              average_precision_score, roc_curve, precision_recall_curve)
from imblearn.over_sampling import SMOTE
from sklearn.compose import ColumnTransformer
from imblearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load Data

```python
df = pd.read_csv('03_BankCustomer_OutlierChecked.csv')
df.head()
```

## 3. Define Features and Target

```python
X = df.drop(columns=['CustomerId', 'Surname', 'Exited'])
y = df['Exited']
```

## 4. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## 5. Preprocessing

```python
num_features = ["CreditScore", "Age", "EstimatedSalary", "Balance", "NumOfProducts", "Tenure"]
cat_features = ["Geography", "Gender", "HasCrCard", "IsActiveMember"]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_features)
    ]
)
```

## 6. Build the Pipeline with SMOTE

```python
steps = [("preprocess", preprocessor),
         ("smote", SMOTE(random_state=42)),
         ("random_forest", RandomForestClassifier(random_state=42))]

pipe = Pipeline(steps)
pipe
```

## 7. Baseline: Untuned Forest + SMOTE

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
y_pred
```

## 8. Hyperparameter Grid

```python
param_grid = {
    'random_forest__n_estimators': [100, 200, 300],
    'random_forest__max_depth': [10, 20, None],
    'random_forest__min_samples_split': [2, 5, 10],
    'random_forest__min_samples_leaf': [1, 2, 4],
    'random_forest__class_weight': [None, 'balanced']
}
```

## 9. Run Grid Search

```python
grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5,
    scoring='recall',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

## 10. Predict Using the Best Model

```python
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)
y_pred
```

## 11. Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Churn", "Churn"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix — Random Forest + SMOTE")
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 12. Per-Class Recall

```python
class0_recall = cm[0, 0] / cm[0].sum()
class1_recall = cm[1, 1] / cm[1].sum()

print(f"Class 0 (Not Churn) Recall: {class0_recall:.2f}")
print(f"Class 1 (Churn) Recall: {class1_recall:.2f}")
```

## 13. Overall Metrics

```python
print(f'Accuracy  : {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision : {precision_score(y_test, y_pred):.2f}')
print(f'Recall    : {recall_score(y_test, y_pred):.2f}')
print(f'F1 Score  : {f1_score(y_test, y_pred):.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

## 14. Predicted Probabilities and Threshold-Independent Metrics

```python
y_prob = best_rf.predict_proba(X_test)[:, 1]
y_prob
```

```python
roc_auc = roc_auc_score(y_test, y_prob)
pr_auc = average_precision_score(y_test, y_prob)
print(f"ROC-AUC Score: {roc_auc:.2f}")
print(f"PR-AUC Score : {pr_auc:.2f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, linewidth=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--", linewidth=1)
plt.xlabel("False Positive Rate", fontsize=12)
plt.ylabel("True Positive Rate", fontsize=12)
plt.title("ROC Curve — Random Forest + SMOTE", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(recall, precision, linewidth=2, marker=".", label=f"PR Curve (AUC = {pr_auc:.2f})")
plt.xlabel("Recall", fontsize=12)
plt.ylabel("Precision", fontsize=12)
plt.title("Precision-Recall Curve — Random Forest + SMOTE", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## 15. Comparison: All Random Forest Variants So Far *(added)*

```python
comparison = pd.DataFrame({
    "RF (plain, class_weight)": [0.85, 0.61, 0.68, 0.65, 0.88, 0.71],
    "RF + UnderSampling":       [0.79, 0.49, 0.78, 0.60, 0.87, 0.69],
    "RF + SMOTE": [
        round(accuracy_score(y_test, y_pred), 2),
        round(precision_score(y_test, y_pred), 2),
        round(recall_score(y_test, y_pred), 2),
        round(f1_score(y_test, y_pred), 2),
        round(roc_auc, 2),
        round(pr_auc, 2),
    ]
}, index=["Accuracy", "Precision", "Recall", "F1 Score", "ROC-AUC", "PR-AUC"])

comparison
```

## 16. Summary

- **Best precision of any Random Forest variant tested (~0.64)**, but the **lowest
  recall (~0.60)** — the opposite trade-off profile from UnderSampling.
- **Why this might happen:** SMOTE's synthetic points are smooth interpolations in
  scaled feature space. When many trees (each seeing a different random subset of this
  data) vote together, the model appears to become more conservative about flagging
  churn — it agrees "churn" only when more confident, boosting precision at recall's
  expense. This is a different interaction than UnderSampling, which simply exposes
  every tree to a smaller, perfectly balanced set of real examples.
- **Emerging picture across three Random Forest variants:**

| Variant | Precision | Recall | Best for... |
|---|---|---|---|
| Plain (class_weight) | 0.61 | 0.68 | Balanced overall performance |
| UnderSampling | 0.49 | 0.78 | Catching the most churners, tolerating false alarms |
| SMOTE | 0.64 | 0.60 | Minimizing false alarms, accepting some missed churners |

There's a genuine, usable trade-off curve here — unlike the Decision Tree series, where
one technique (`class_weight`) simply won outright. The next notebook (ADASYN) will
determine whether it lands close to SMOTE (its origin technique) or carves out its own
position on this trade-off curve.
