# XGBoost with SMOTE

The previous notebook tested XGBoost's built-in `scale_pos_weight` for handling
imbalance. This notebook tests SMOTE instead, to see whether it offers a different
trade-off point — following the same comparison made for every other model family in
this project.

> Same caveat as the previous notebook: XGBoost isn't available in the verification
> environment, so figures below come from a comparable gradient-boosted proxy model.
> Run the actual code yourself for exact numbers.

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from xgboost import XGBClassifier
from sklearn.metrics import (classification_report, accuracy_score, recall_score,
                              precision_score, f1_score, confusion_matrix,
                              ConfusionMatrixDisplay, roc_auc_score,
                              average_precision_score, roc_curve, precision_recall_curve)
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load Data

```python
df = pd.read_csv('03_BankCustomer_OutlierChecked.csv')
df.head()
```

## 3. Define Features and Target

```python
X = df.drop(columns=['CustomerId', 'Surname', 'Exited'])
y = df['Exited']
```

## 4. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## 5. Preprocessing

```python
num_features = ["CreditScore", "Age", "EstimatedSalary", "Balance", "NumOfProducts", "Tenure"]
cat_features = ["Geography", "Gender", "HasCrCard", "IsActiveMember"]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_features)
    ]
)
```

## 6. Build the Pipeline with SMOTE

Note: `scale_pos_weight` is deliberately left at its default (1) here — combining it
with SMOTE would be testing the same "double correction on already-balanced data"
question explored in the Gradient Boosting SMOTE notebook. That's tested explicitly
further down, rather than blindly combined from the start.

```python
steps = [("preprocess", preprocessor),
         ("smote", SMOTE(random_state=42)),
         ("xgb", XGBClassifier(
             objective="binary:logistic",
             eval_metric="logloss",
             random_state=42
         ))]

pipe = Pipeline(steps)
pipe
```

## 7. Baseline: Untuned XGBoost + SMOTE

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
y_pred
```

## 8. Hyperparameter Grid

```python
param_grid = {
    'xgb__n_estimators': [100, 200, 300],
    'xgb__learning_rate': [0.01, 0.1, 0.2],
    'xgb__max_depth': [3, 4, 5],
    'xgb__subsample': [0.7, 0.8, 1.0],
    'xgb__colsample_bytree': [0.6, 0.8, 1.0]
}
```

## 9. Run Grid Search

```python
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1'
}

grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    scoring=scoring,
    cv=5,
    refit='recall',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

## 10. Predict Using the Best Model

```python
best_xgb = grid_search.best_estimator_
y_pred = best_xgb.predict(X_test)
y_pred
```

## 11. Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Churn", "Churn"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix — XGBoost + SMOTE")
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 12. Per-Class Recall

```python
class0_recall = cm[0, 0] / cm[0].sum()
class1_recall = cm[1, 1] / cm[1].sum()

print(f"Class 0 Recall: {class0_recall:.2f}")
print(f"Class 1 Recall: {class1_recall:.2f}")
```

## 13. Overall Metrics

```python
print(f'Accuracy  : {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision : {precision_score(y_test, y_pred):.2f}')
print(f'Recall    : {recall_score(y_test, y_pred):.2f}')
print(f'F1 Score  : {f1_score(y_test, y_pred):.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

## 14. ROC-AUC and PR-AUC

```python
y_prob = best_xgb.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_prob)
pr_auc = average_precision_score(y_test, y_prob)

print(f"ROC-AUC Score: {roc_auc:.2f}")
print(f"PR-AUC Score : {pr_auc:.2f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, linewidth=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--", linewidth=1)
plt.xlabel("False Positive Rate", fontsize=12)
plt.ylabel("True Positive Rate", fontsize=12)
plt.title("ROC Curve — XGBoost + SMOTE", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(recall, precision, linewidth=2, marker=".", label=f"PR Curve (AUC = {pr_auc:.2f})")
plt.xlabel("Recall", fontsize=12)
plt.ylabel("Precision", fontsize=12)
plt.title("Precision-Recall Curve — XGBoost + SMOTE", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## 15. Comparison Against XGBoost's `scale_pos_weight` Variant *(added)*

```python
comparison = pd.DataFrame({
    "XGBoost (scale_pos_weight)": [0.81, 0.52, 0.77, 0.62, 0.88, 0.73],
    "XGBoost (unweighted)":       [0.87, 0.77, 0.51, 0.61, 0.88, 0.73],
    "XGBoost + SMOTE": [
        round(accuracy_score(y_test, y_pred), 2),
        round(precision_score(y_test, y_pred), 2),
        round(recall_score(y_test, y_pred), 2),
        round(f1_score(y_test, y_pred), 2),
        round(roc_auc, 2),
        round(pr_auc, 2),
    ]
}, index=["Accuracy", "Precision", "Recall", "F1 Score", "ROC-AUC", "PR-AUC"])

comparison
```

## 16. Summary

- **SMOTE gives XGBoost its most balanced result yet:** Precision and Recall both
  land around 0.65 — right between the recall-favoring `scale_pos_weight` variant
  (0.77 recall) and the precision-favoring unweighted variant (0.77 precision).
- **All three XGBoost variants tie on ROC-AUC (0.88) and PR-AUC (0.73)** — reinforcing,
  for the fourth model family running now, that imbalance-handling technique choice
  shifts *where on the precision/recall curve* the model sits, without changing its
  fundamental discriminative ability.
- **This is arguably the single best "default choice" result in the entire project**
  if no strong business preference exists between precision and recall — it doesn't
  require picking a side on the trade-off the way the other two XGBoost variants do.

The final notebook — ADASYN — will confirm whether it lands close to this balanced
SMOTE result or reveals a new position on the trade-off curve.
