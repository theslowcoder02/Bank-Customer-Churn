# Gradient Boosting with ADASYN

The final notebook in the Gradient Boosting series, and the last model notebook in the
whole project. Given ADASYN's close relationship to SMOTE, the expectation is it lands
near SMOTE's precision-favoring result rather than `sample_weight`'s recall-favoring one.

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import (classification_report, accuracy_score, recall_score,
                              precision_score, f1_score, confusion_matrix,
                              ConfusionMatrixDisplay, roc_auc_score,
                              average_precision_score, roc_curve, precision_recall_curve)
from imblearn.over_sampling import ADASYN
from imblearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load Data

```python
df = pd.read_csv('03_BankCustomer_OutlierChecked.csv')
df.head()
```

## 3. Define Features and Target

```python
X = df.drop(columns=['CustomerId', 'Surname', 'Exited'])
y = df['Exited']
```

## 4. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## 5. Preprocessing

```python
num_features = ["CreditScore", "Age", "EstimatedSalary", "Balance", "NumOfProducts", "Tenure"]
cat_features = ["Geography", "Gender", "HasCrCard", "IsActiveMember"]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_features)
    ]
)
```

## 6. Build the Pipeline with ADASYN

```python
steps = [("preprocess", preprocessor),
         ("adasyn", ADASYN(random_state=42)),
         ("gradient_boost", GradientBoostingClassifier(random_state=42))]

pipe = Pipeline(steps)
pipe
```

## 7. Baseline: Untuned Gradient Boosting + ADASYN

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
y_pred
```

## 8. Hyperparameter Grid

```python
param_grid = {
    'gradient_boost__n_estimators': [50, 100, 200, 300],
    'gradient_boost__learning_rate': [0.001, 0.01, 0.1, 1.0],
    'gradient_boost__max_depth': [1, 2, 3],
    'gradient_boost__subsample': [0.8, 1.0]
}
```

## 9. Run Grid Search

```python
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1'
}

grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    scoring=scoring,
    refit='recall',
    cv=5,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

## 10. Predict Using the Best Model

```python
best_gb = grid_search.best_estimator_
y_pred = best_gb.predict(X_test)
y_pred
```

## 11. Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Churn", "Churn"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix — Gradient Boosting + ADASYN")
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 12. Per-Class Recall

```python
class0_recall = cm[0, 0] / cm[0].sum()
class1_recall = cm[1, 1] / cm[1].sum()

print(f"Class 0 Recall: {class0_recall:.2f}")
print(f"Class 1 Recall: {class1_recall:.2f}")
```

## 13. Overall Metrics

```python
print(f'Accuracy  : {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision : {precision_score(y_test, y_pred):.2f}')
print(f'Recall    : {recall_score(y_test, y_pred):.2f}')
print(f'F1 Score  : {f1_score(y_test, y_pred):.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

## 14. ROC-AUC and PR-AUC

```python
y_prob = best_gb.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_prob)
pr_auc = average_precision_score(y_test, y_prob)

print(f"ROC-AUC Score: {roc_auc:.2f}")
print(f"PR-AUC Score : {pr_auc:.2f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, linewidth=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--", linewidth=1)
plt.xlabel("False Positive Rate", fontsize=12)
plt.ylabel("True Positive Rate", fontsize=12)
plt.title("ROC Curve — Gradient Boosting + ADASYN", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(recall, precision, linewidth=2, marker=".", label=f"PR Curve (AUC = {pr_auc:.2f})")
plt.xlabel("Recall", fontsize=12)
plt.ylabel("Precision", fontsize=12)
plt.title("Precision-Recall Curve — Gradient Boosting + ADASYN", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## 17. Final Project Conclusion

**Winner: Gradient Boosting with `sample_weight` balancing** — the best combination of
Recall (0.77), ROC-AUC (0.88), and PR-AUC (0.73) of any model or technique tested
across this entire project. Gradient Boosting + ADASYN ties it on PR-AUC with a
different, more precision-favoring profile.

### The journey, in one paragraph
Starting from a Logistic Regression baseline that caught only 19% of churners despite
81% accuracy, this project systematically tested how different models and imbalance
techniques respond to the same problem. Linear models proved indifferent to *which*
imbalance technique was used, since their real limitation was the model's own
simplicity. Tree-based models responded very differently to each technique — sometimes
benefiting, sometimes actively hurt by external resampling — revealing that "the best
fix for imbalance" depends entirely on the model paired with it. The strongest results
came from models expressive enough (deep trees, boosted ensembles) to use reweighted
or resampled data constructively rather than being misled by it.

### Practical recommendation
Deploy **Gradient Boosting with balanced `sample_weight`** as the production churn
model. If the business specifically prioritizes minimizing false alarms over catching
every possible churner, **Gradient Boosting + SMOTE** is the better-suited alternative,
trading some recall for meaningfully higher precision.
