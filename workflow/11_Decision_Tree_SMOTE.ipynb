# Decision Tree with SMOTE

The previous notebook found that combining UnderSampling with a tuned Decision Tree
underperformed the plain tuned tree. This notebook tests the other oversampling
technique from the Logistic Regression series — SMOTE — to see whether generating
synthetic churn examples (instead of deleting real majority data) fares any better with
a tree-based model.

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (classification_report, accuracy_score, precision_score,
                              f1_score, recall_score, confusion_matrix,
                              ConfusionMatrixDisplay, roc_auc_score,
                              average_precision_score, roc_curve, precision_recall_curve)
from imblearn.over_sampling import SMOTE
from sklearn.compose import ColumnTransformer
from imblearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load Data

```python
df = pd.read_csv('03_BankCustomer_OutlierChecked.csv')
df.head()
```

## 3. Define Features and Target

```python
X = df.drop(columns=['CustomerId', 'Surname', 'Exited'])
y = df['Exited']
```

## 4. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## 5. Preprocessing

```python
num_features = ["CreditScore", "Age", "EstimatedSalary", "Balance", "NumOfProducts", "Tenure"]
cat_features = ["Geography", "Gender", "HasCrCard", "IsActiveMember"]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_features)
    ]
)
```

## 6. Build the Pipeline with SMOTE

Same reasoning as the Logistic Regression SMOTE notebook: SMOTE runs after
preprocessing (so distances between points are measured on scaled, numeric data), and
only ever touches the training folds via `imblearn`'s `Pipeline`.

```python
steps = [("preprocess", preprocessor),
         ("SMOTE", SMOTE(random_state=42)),
         ("decision_tree", DecisionTreeClassifier(random_state=42))]

pipe = Pipeline(steps)
pipe
```

## 7. Baseline: Untuned Tree + SMOTE

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
y_pred
```

## 8. Hyperparameter Grid

Same grid as both previous Decision Tree notebooks, including `class_weight` — testing
here whether stacking class weighting on top of *synthetic* oversampling causes the
same kind of over-correction problem seen with UnderSampling.

```python
param_grid = {
    'decision_tree__criterion': ['gini', 'entropy'],
    'decision_tree__max_depth': [None, 10, 20, 30],
    'decision_tree__min_samples_split': [2, 5, 10],
    'decision_tree__min_samples_leaf': [1, 2, 4],
    'decision_tree__class_weight': [None, 'balanced']
}
```

## 9. Run Grid Search

```python
grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5,
    scoring='recall',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

## 10. Predict Using the Best Model

```python
best_dt = grid_search.best_estimator_
y_pred = best_dt.predict(X_test)
y_pred
```

## 11. Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Churn", "Churn"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix — Decision Tree + SMOTE")
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 12. Per-Class Recall

```python
class0_recall = cm[0, 0] / cm[0].sum()
class1_recall = cm[1, 1] / cm[1].sum()

print(f"Class 0 (Not Churn) Recall: {class0_recall:.2f}")
print(f"Class 1 (Churn) Recall: {class1_recall:.2f}")
```

## 13. Overall Metrics

```python
print(f'Accuracy  : {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision : {precision_score(y_test, y_pred):.2f}')
print(f'Recall    : {recall_score(y_test, y_pred):.2f}')
print(f'F1 Score  : {f1_score(y_test, y_pred):.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

## 14. Predicted Probabilities and Threshold-Independent Metrics

```python
y_prob = best_dt.predict_proba(X_test)[:, 1]
y_prob
```

```python
roc_auc = roc_auc_score(y_test, y_prob)
pr_auc = average_precision_score(y_test, y_prob)
print(f"ROC-AUC Score: {roc_auc:.2f}")
print(f"PR-AUC Score : {pr_auc:.2f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, linewidth=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--", linewidth=1)
plt.xlabel("False Positive Rate", fontsize=12)
plt.ylabel("True Positive Rate", fontsize=12)
plt.title("ROC Curve — Decision Tree + SMOTE", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(recall, precision, linewidth=2, marker=".", label=f"PR Curve (AUC = {pr_auc:.2f})")
plt.xlabel("Recall", fontsize=12)
plt.ylabel("Precision", fontsize=12)
plt.title("Precision-Recall Curve — Decision Tree + SMOTE", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## 15. Comparison Across All Decision Tree Variants *(added)*

```python
comparison = pd.DataFrame({
    "Plain Tuned Tree":       [0.73, 0.41, 0.74, 0.52, 0.79, 0.59],
    "Tree + UnderSampling":   [0.71, 0.39, 0.70, 0.50, 0.76, 0.40],
    "Tree + SMOTE": [
        round(accuracy_score(y_test, y_pred), 2),
        round(precision_score(y_test, y_pred), 2),
        round(recall_score(y_test, y_pred), 2),
        round(f1_score(y_test, y_pred), 2),
        round(roc_auc, 2),
        round(pr_auc, 2),
    ]
}, index=["Accuracy", "Precision", "Recall", "F1 Score", "ROC-AUC", "PR-AUC"])

comparison
```

## 16. Summary

- **Worse across nearly every metric** compared to the plain tuned tree: Recall drops
  from ~0.74 to ~0.59, ROC-AUC drops from ~0.79 to ~0.71, and PR-AUC drops sharply from
  ~0.59 to ~0.37.
- **Why SMOTE likely hurts tree-based models specifically:** SMOTE creates synthetic
  points by interpolating *between* real minority points in continuous feature space.
  Decision Trees split on axis-aligned thresholds (e.g. "Age > 45"), not smooth
  boundaries — synthetic points invented via interpolation can sit in awkward positions
  that encourage the tree to create overly specific, noisy splits chasing patterns in
  fabricated data rather than the real distribution. This is a known limitation of
  combining SMOTE with tree-based models, in contrast to Logistic Regression (a smooth,
  linear model), which handled interpolated synthetic points without issue.
- **Reinforces the finding from the UnderSampling notebook:** once a Decision Tree
  already has `class_weight` available as a built-in, tunable imbalance-handling
  mechanism, layering an external resampling technique on top tends to hurt rather than
  help — for a different underlying reason each time (UnderSampling removes real data;
  SMOTE introduces awkward synthetic geometry that trees are especially sensitive to).

**Running conclusion so far:** across two tree-based experiments, the plain tuned
Decision Tree with `class_weight='balanced'` remains the strongest model in this
project. The next notebook (Class Weighting) is expected to essentially recover that
same result, since it's the exact mechanism already embedded in the winning
configuration here — worth confirming directly rather than assuming.
