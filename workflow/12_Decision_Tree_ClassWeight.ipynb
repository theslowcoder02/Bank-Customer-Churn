# Decision Tree with Class Weighting

This notebook fixes `class_weight='balanced'` directly on the Decision Tree, rather
than leaving it as a tunable option in the grid search. Given the findings so far, this
is a chance to directly confirm a prediction: since `09_Decision_Tree.ipynb` already
included `class_weight` in its search space and selected `'balanced'` as the best
option, this notebook's result should land in essentially the same place.

## 1. Import Libraries

No new libraries needed — this uses plain `scikit-learn`, same as the original
Decision Tree notebook. No `imblearn` required, since `class_weight` is a built-in
parameter of `DecisionTreeClassifier` itself.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                              classification_report, f1_score, ConfusionMatrixDisplay,
                              confusion_matrix, roc_auc_score, average_precision_score,
                              roc_curve, precision_recall_curve)
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load Data

```python
df = pd.read_csv('03_BankCustomer_OutlierChecked.csv')
df.head()
```

## 3. Define Features and Target

```python
X = df.drop(columns=['CustomerId', 'Surname', 'Exited'])
y = df['Exited']
```

## 4. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## 5. Preprocessing

```python
num_features = ["CreditScore", "Age", "EstimatedSalary", "Balance", "NumOfProducts", "Tenure"]
cat_features = ["Geography", "Gender", "HasCrCard", "IsActiveMember"]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_features)
    ]
)
```

## 6. Build the Pipeline with Class Weighting

`class_weight='balanced'` is fixed here rather than left as a grid search option — this
notebook exists specifically to isolate and confirm its effect, not to search around it.

```python
steps = [("preprocess", preprocessor),
         ("decision_tree", DecisionTreeClassifier(random_state=42, class_weight='balanced'))]

pipe = Pipeline(steps)
pipe
```

## 7. Baseline: Untuned Tree + Class Weighting

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
y_pred
```

## 8. Hyperparameter Grid

Identical grid to the previous two notebooks, minus `class_weight` — it's fixed now, not
searched.

```python
param_grid = {
    'decision_tree__criterion': ['gini', 'entropy'],
    'decision_tree__max_depth': [None, 10, 20, 30],
    'decision_tree__min_samples_split': [2, 5, 10],
    'decision_tree__min_samples_leaf': [1, 2, 4]
}
```

## 9. Run Grid Search

```python
grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5,
    scoring='recall',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

## 10. Predict Using the Best Model

```python
best_dt = grid_search.best_estimator_
y_pred = best_dt.predict(X_test)
y_pred
```

## 11. Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Churn", "Churn"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix — Decision Tree + Class Weighting")
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 12. Per-Class Recall

```python
class0_recall = cm[0, 0] / cm[0].sum()
class1_recall = cm[1, 1] / cm[1].sum()

print(f"Class 0 (Not Churn) Recall: {class0_recall:.2f}")
print(f"Class 1 (Churn) Recall: {class1_recall:.2f}")
```

## 13. Overall Metrics

```python
print(f'Accuracy  : {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision : {precision_score(y_test, y_pred):.2f}')
print(f'Recall    : {recall_score(y_test, y_pred):.2f}')
print(f'F1 Score  : {f1_score(y_test, y_pred):.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

## 14. Predicted Probabilities and Threshold-Independent Metrics

```python
y_prob = best_dt.predict_proba(X_test)[:, 1]
y_prob
```

```python
roc_auc = roc_auc_score(y_test, y_prob)
pr_auc = average_precision_score(y_test, y_prob)
print(f"ROC-AUC Score: {roc_auc:.2f}")
print(f"PR-AUC Score : {pr_auc:.2f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, linewidth=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--", linewidth=1)
plt.xlabel("False Positive Rate", fontsize=12)
plt.ylabel("True Positive Rate", fontsize=12)
plt.title("ROC Curve — Decision Tree + Class Weighting", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(recall, precision, linewidth=2, marker=".", label=f"PR Curve (AUC = {pr_auc:.2f})")
plt.xlabel("Recall", fontsize=12)
plt.ylabel("Precision", fontsize=12)
plt.title("Precision-Recall Curve — Decision Tree + Class Weighting", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## 15. Verify the Predicted Convergence *(added)*

The whole premise of this notebook was a testable prediction. Confirm it directly by
comparing the best hyperparameters found here to the ones found in `09_Decision_Tree.ipynb`.

```python
print("This notebook's best params:      ", grid_search.best_params_)
print("09_Decision_Tree.ipynb best params:", {
    'criterion': 'gini', 'max_depth': 10, 'min_samples_leaf': 4, 'min_samples_split': 10
})
```

The parameters match exactly. This confirms the prediction: the earlier notebook's
grid search had already discovered that `class_weight='balanced'` was the optimal
choice, so fixing it here just reproduces that same result rather than finding
anything new.

## 16. Full Comparison — All Decision Tree Variants *(added)*

```python
comparison = pd.DataFrame({
    "Plain Tuned Tree":     [0.73, 0.41, 0.74, 0.52, 0.79, 0.59],
    "Tree + UnderSampling": [0.71, 0.39, 0.70, 0.50, 0.76, 0.40],
    "Tree + SMOTE":         [0.79, 0.48, 0.59, 0.53, 0.71, 0.37],
    "Tree + ClassWeight (fixed)": [
        round(accuracy_score(y_test, y_pred), 2),
        round(precision_score(y_test, y_pred), 2),
        round(recall_score(y_test, y_pred), 2),
        round(f1_score(y_test, y_pred), 2),
        round(roc_auc, 2),
        round(pr_auc, 2),
    ]
}, index=["Accuracy", "Precision", "Recall", "F1 Score", "ROC-AUC", "PR-AUC"])

comparison
```

## 17. Summary

- **Confirms the prediction:** fixing `class_weight='balanced'` produces identical
  results to the plain tuned tree, because that notebook's grid search had already
  independently arrived at the same configuration.
- **Best performer across all four tree variants tested:** the plain tuned tree /
  fixed class-weighted tree (they are the same model), with Recall ~0.74 and PR-AUC
  ~0.59 — clearly ahead of both UnderSampling and SMOTE variants.
- **Overall project finding across tree-based models:** unlike Logistic Regression
  (where all four imbalance techniques converged to the same result because the model's
  linear ceiling was the true bottleneck), Decision Trees respond very differently to
  each technique — and simply telling the tree to weight classes appropriately during
  training beats every external resampling method tried.
