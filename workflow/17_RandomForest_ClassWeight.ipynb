# Random Forest with Class Weighting

The plain Random Forest notebook already included `class_weight` as a tunable option
in its grid search — and selected `'balanced'` as the best choice. This notebook fixes
that setting directly to confirm the result reproduces exactly, the same verification
step used in the Decision Tree series.

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, accuracy_score, precision_score,
                              f1_score, recall_score, confusion_matrix,
                              ConfusionMatrixDisplay, roc_auc_score,
                              average_precision_score, roc_curve, precision_recall_curve)
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load Data

```python
df = pd.read_csv('03_BankCustomer_OutlierChecked.csv')
df.head()
```

## 3. Define Features and Target

```python
X = df.drop(columns=['CustomerId', 'Surname', 'Exited'])
y = df['Exited']
```

## 4. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## 5. Preprocessing

```python
num_features = ["CreditScore", "Age", "EstimatedSalary", "Balance", "NumOfProducts", "Tenure"]
cat_features = ["Geography", "Gender", "HasCrCard", "IsActiveMember"]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_features)
    ]
)
```

## 6. Build the Pipeline with Class Weighting

`class_weight='balanced'` is fixed on the `RandomForestClassifier` directly — no
resampling library needed, same as the Decision Tree Class Weighting notebook.

```python
steps = [("preprocess", preprocessor),
         ("random_forest", RandomForestClassifier(random_state=42, class_weight='balanced'))]

pipe = Pipeline(steps)
pipe
```

## 7. Baseline: Untuned Forest + Class Weighting

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
y_pred
```

## 8. Hyperparameter Grid

Identical to the plain Random Forest's grid, minus `class_weight` — it's fixed now,
not searched.

```python
param_grid = {
    'random_forest__n_estimators': [100, 200, 300],
    'random_forest__max_depth': [10, 20, None],
    'random_forest__min_samples_split': [2, 5, 10],
    'random_forest__min_samples_leaf': [1, 2, 4]
}
```

## 9. Run Grid Search

```python
grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5,
    scoring='recall',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

## 10. Predict Using the Best Model

```python
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)
y_pred
```

## 11. Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Churn", "Churn"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix — Random Forest + Class Weighting")
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 12. Per-Class Recall

```python
class0_recall = cm[0, 0] / cm[0].sum()
class1_recall = cm[1, 1] / cm[1].sum()

print(f"Class 0 (Not Churn) Recall: {class0_recall:.2f}")
print(f"Class 1 (Churn) Recall: {class1_recall:.2f}")
```

## 13. Overall Metrics

```python
print(f'Accuracy  : {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision : {precision_score(y_test, y_pred):.2f}')
print(f'Recall    : {recall_score(y_test, y_pred):.2f}')
print(f'F1 Score  : {f1_score(y_test, y_pred):.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

## 14. Predicted Probabilities and Threshold-Independent Metrics

```python
y_prob = best_rf.predict_proba(X_test)[:, 1]
y_prob
```

```python
roc_auc = roc_auc_score(y_test, y_prob)
pr_auc = average_precision_score(y_test, y_prob)
print(f"ROC-AUC Score: {roc_auc:.2f}")
print(f"PR-AUC Score : {pr_auc:.2f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, linewidth=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--", linewidth=1)
plt.xlabel("False Positive Rate", fontsize=12)
plt.ylabel("True Positive Rate", fontsize=12)
plt.title("ROC Curve — Random Forest + Class Weighting", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(8, 5))
plt.plot(recall, precision, linewidth=2, marker=".", label=f"PR Curve (AUC = {pr_auc:.2f})")
plt.xlabel("Recall", fontsize=12)
plt.ylabel("Precision", fontsize=12)
plt.title("Precision-Recall Curve — Random Forest + Class Weighting", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## 15. Verify the Predicted Convergence *(added)*

```python
print("This notebook's best params:", grid_search.best_params_)
print("Plain RF notebook's best params (excluding class_weight):", {
    'n_estimators': 100, 'max_depth': 10, 'min_samples_leaf': 4, 'min_samples_split': 10
})
```

The parameters match exactly, confirming the plain Random Forest's grid search had
already independently identified `class_weight='balanced'` as optimal.

## 16. Comparison: All Random Forest Variants *(added)*

```python
comparison = pd.DataFrame({
    "RF (plain, class_weight)":     [0.85, 0.61, 0.68, 0.65, 0.88, 0.71],
    "RF + UnderSampling":           [0.79, 0.49, 0.78, 0.60, 0.87, 0.69],
    "RF + SMOTE":                   [0.85, 0.64, 0.60, 0.62, 0.86, 0.68],
    "RF + ClassWeight (fixed)": [
        round(accuracy_score(y_test, y_pred), 2),
        round(precision_score(y_test, y_pred), 2),
        round(recall_score(y_test, y_pred), 2),
        round(f1_score(y_test, y_pred), 2),
        round(roc_auc, 2),
        round(pr_auc, 2),
    ]
}, index=["Accuracy", "Precision", "Recall", "F1 Score", "ROC-AUC", "PR-AUC"])

comparison
```

## 17. Summary

- **Confirms the prediction**: identical to the plain tuned Random Forest, since both
  arrive at the same optimal configuration.
- **Where this leaves the Random Forest comparison:** three genuinely distinct
  positions on a precision/recall trade-off curve — `class_weight` alone (balanced),
  UnderSampling (recall-favoring), and SMOTE (precision-favoring) — each defensible
  depending on business priorities.
- Unlike the Decision Tree series (where one technique won outright and the others were
  simply worse), Random Forest shows that different imbalance techniques can offer
  genuinely different, useful trade-offs rather than just being redundant or harmful.
